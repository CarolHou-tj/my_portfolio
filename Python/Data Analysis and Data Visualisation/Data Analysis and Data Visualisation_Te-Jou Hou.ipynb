```python
A = len(df[df['age'] <= 45])
B = len(df[df['age'] > 45])
C = int(df['age'].min())
D = int(df['age'].max())
E = len(df[df['DEATH_EVENT'] == 1])
F = len(df)
G = F - E
H = round(float((E / F) * 100), 2)
I = round(float((df[(df['age'] <= 45) & (df['DEATH_EVENT'] == 1)].shape[0] / A) * 100), 2)
J = round(float((df[(df['age'] > 45) & (df['DEATH_EVENT'] == 1)].shape[0] / B) * 100), 2)
K = round(float((df[(df['smoking'] == 1) & (df['DEATH_EVENT'] == 1)].shape[0] / df[df['smoking'] == 1].shape[0]) * 100), 2) 
L = df[df['sex'] == 1].shape[0]
M = df[(df['sex'] == 1) & (df['DEATH_EVENT'] == 0)].shape[0]
N = round(float((M / L) * 100), 2)
O = df[df['sex'] == 0].shape[0]
P = df[(df['sex'] == 0) & (df['DEATH_EVENT'] == 0)].shape[0]
Q = round(float((P / O) * 100), 2)

print(f"The number of people 45 or less in the database is {A}.")
print(f"The number of people older than 45 in the database is {B}.")
print(f"The youngest person in the database is {C} years old.")
print(f"The oldest person in the database is {D} years old.")
print("\n")
print(f"The number of people in the database who died following a heart failure event is {E} out of {F} patients.")
print(f"The number of people in the database who survived following a heart failure event: {G} out of {F} patients.")
print("\n")
print(f"In other words, the percentage of people who died following a heart failure event is {H}%.")
print("\n")
print(f"The percentage of young people (<= 45) who died after a heart failure event is {I}%.")
print(f"The percentage of old people (> 45) who died after a heart failure event is {J}%.")
print(f"The percentage of smokers who died after a heart failure event is {K}%.")
print("\n")
print(f"The number of men in the database is {L}.")
print(f"The number of men survivors in the database is {M} or {N}%.")
print(f"The number of women in the database is {O}.")
print(f"The number of women survivors in the database is {P} or {Q}%.")
```

```python
from matplotlib import pyplot as plt
```

```python
def func(s,d):
  t = int(round(s/100.*sum(d)))
  return f'{s:.1f}%({t})' 


plt.subplot(1,3,1) #subplpot#1

bins = [0, 50, 60, 70, 80, 90, 100]
labels = ['40s', '50s', '60s', '70s', '80s', '90s']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
age_counts = df['age_group'].value_counts(sort=False)

plt.pie(age_counts, labels=labels, autopct=lambda i: func(i, age_counts), radius = 25, textprops={'fontsize': 6})
plt.title('Age of Patients who suffered heart failure', fontsize=6.5, pad=50)
 
    
plt.subplot(1,3,2) #subplot#2

sex_counts = df['sex'].value_counts()
sex_labels = ['Female', 'Male']
sex_counts = sex_counts.sort_index(ascending=True)

plt.pie(sex_counts, labels=sex_labels, autopct=lambda i: func(i, sex_counts), radius = 25, textprops={'fontsize': 6})
plt.title('Proportion of Male/Female who experienced Heart Failure', fontsize=6.5, pad=50)


plt.subplot(1,3,3) #subplot#3

df_survived = df['DEATH_EVENT'] == 0
survived_counts = df_survived.value_counts()
survived_labels = ['Died', 'Survived']
survived_counts = survived_counts.sort_index(ascending=True)

plt.pie(survived_counts, labels=survived_labels, autopct=lambda i: func(i, survived_counts), radius = 25, textprops={'fontsize': 6})
plt.title('Heart Failure Survivors', fontsize=6.5, pad=50)


plt.subplots_adjust(wspace=50)
           
plt.show()
```

```python
sex_counts = df[df['DEATH_EVENT'] == 0]['sex'].value_counts()
sex_labels = ['Male', 'Female']
sex_counts = sex_counts.sort_index(ascending=True)

plt.pie(sex_counts, labels=sex_labels, autopct=lambda i: func(i, sex_counts))
plt.title('Proportion of Male/Female Heart Failure Survivors')

plt.show()
```

```python
survivors = df[df['DEATH_EVENT'] == 0]
sex_survivors = survivors['sex'].value_counts()
sex_total = df['sex'].value_counts()
sex_survival_percentage = round((sex_survivors / sex_total) * 100, 2)
sex_survival_percentage.index = ['Male', 'Female']

bars = plt.bar(sex_survival_percentage.index, sex_survival_percentage, color = 'green')
plt.title('Percentage of survivors per Gender')
plt.xlabel("Gender")
plt.ylabel("Percentage of survivors")
for i, bar in enumerate(bars):
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval / 2, 
             f'{sex_survival_percentage[i]}', ha='center', va='center', fontsize=10, color = 'white')

plt.show()
```

```python
survivors = df[df['DEATH_EVENT'] == 0].copy()
bins = [0, 50, 60, 70, 80, 90, 100]
labels = ['40s', '50s', '60s', '70s', '80s', '90s']
survivors.loc[:, 'age_group'] = pd.cut(survivors['age'], bins=bins, labels=labels, right=False)
age_group_counts = survivors['age_group'].value_counts(sort=False)
total_survivors = len(survivors)
age_group_survival_percentage = (age_group_counts / total_survivors) * 100

plt.plot(age_group_survival_percentage.index, age_group_survival_percentage)
plt.title('Percentage of Survivors per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Survivors')
plt.show()
```
